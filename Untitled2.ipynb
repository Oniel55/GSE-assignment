```python
# Dataset Management and Basic Analysis System

class DataSet:
    def __init__(self, numeric_file, category_file, threshold=50):
        self.numeric_file = numeric_file
        self.category_file = category_file
        self.threshold = threshold
        self.data = []
        self.categories = set()
        self.total = 0
        self.average = 0
        self.minimum = None
        self.maximum = None

    # 1 & 2. File Handling + Error Handling
    def load_data(self):
        # Load numerical data
        try:
            with open(self.numeric_file, 'r') as f:
                for line in f:
                    line = line.strip()
                    if line == "":
                        continue
                    try:
                        value = float(line)
                        self.data.append(value)
                    except ValueError:
                        print(f"Invalid value ignored: {line}")

            if len(self.data) == 0:
                raise ValueError("Numeric file is empty or contains no valid data")

        except FileNotFoundError:
            print("Error: Numeric data file not found.")
            return False
        except ValueError as e:
            print(f"Error: {e}")
            return False

        # Load categorical data
        try:
            with open(self.category_file, 'r') as f:
                for line in f:
                    line = line.strip()
                    if line:
                        self.categories.add(line)
        except FileNotFoundError:
            print("Error: Category file not found.")
            return False

        return True

    # 3 & 4. Functions, Operators, and Loops
    def calculate_statistics(self):
        total = 0
        count = 0
        minimum = self.data[0]
        maximum = self.data[0]

        for value in self.data:
            total += value
            count += 1

            if value < minimum:
                minimum = value
            if value > maximum:
                maximum = value

        average = total / count

        self.total = total
        self.average = average
        self.minimum = minimum
        self.maximum = maximum

    # 5 & 6. Conditionals and Sets
    def display_results(self):
        print("----- Analysis Results -----")
        print(f"Total: {self.total}")
        print(f"Average: {self.average}")
        print(f"Minimum: {self.minimum}")
        print(f"Maximum: {self.maximum}")

        if self.average > self.threshold:
            print("Performance: High Performance")
        else:
            print("Performance: Needs Improvement")

        print(f"Unique Categories: {len(self.categories)}")

    # 8. Save results to file
    def save_report(self, report_file):
        with open(report_file, 'w') as f:
            f.write("Dataset Analysis Report\n")
            f.write("-----------------------\n")
            f.write(f"Total: {self.total}\n")
            f.write(f"Average: {self.average}\n")
            f.write(f"Minimum: {self.minimum}\n")
            f.write(f"Maximum: {self.maximum}\n")

            if self.average > self.threshold:
                f.write("Performance: High Performance\n")
            else:
                f.write("Performance: Needs Improvement\n")

            f.write(f"Unique Categories Count: {len(self.categories)}\n")


# 7. Create object and run analysis
dataset = DataSet("data.txt", "categories.txt", threshold=50)

if dataset.load_data():
    dataset.calculate_statistics()
    dataset.display_results()
    dataset.save_report("report.txt")
    print("\nReport saved successfully.")
```
